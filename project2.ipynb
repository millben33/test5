```python
import psycopg2
import pandas as pd
import sys
import pandas.io.sql as psql
from datetime import datetime
import pytz
# Connection parameters, yours will be different

param_dic = {
    "host"      : "localhost",
    "database"  : "sql_data",
    "user"      : "ben1",
    "password"  : "123456"
}
class data_source:
    def __init__(self,**data):
        self.data=data;
        conn_detail="To be connect {host} and {database} with {user} and {password}"
        self.conn=None;
        self.d=1;
        print(conn_detail.format(**data))
        self.connect()
    def connect(self):
      """ Connect to the PostgreSQL database server """
     # conn = None
      try:
        # connect to the PostgreSQL server
        print('Connecting to the PostgreSQL database...')
        self.conn = psycopg2.connect(**self.data)
      except (Exception, psycopg2.DatabaseError) as error:
        print(error)
        sys.exit(1) 
      self.d=2
      print("Connection successful",self.d)
      return self.conn
    def join_all_table(self,sql_s):
         print(sql_s,"self.conn",self.conn,self.d)
         df=psql.read_sql(sql_s,self.conn)
         return df
    def make_new_df (self,df,column_name,value):
        df_new=df[df[column_name]==value].copy()
        return df_new
```
